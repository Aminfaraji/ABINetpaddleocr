```python
!unrar x /content/drive/MyDrive/final_dataset_idcart.rar
import shutil
from glob import glob
!mkdir /content/yolov9/final_dataset_idcart/train
!mkdir /content/yolov9/final_dataset_idcart/test
!mkdir /content/yolov9/final_dataset_idcart/val
!mkdir /content/yolov9/final_dataset_idcart/train/images
!mkdir /content/yolov9/final_dataset_idcart/train/labels
!mkdir /content/yolov9/final_dataset_idcart/test/images
!mkdir /content/yolov9/final_dataset_idcart/test/labels
!mkdir /content/yolov9/final_dataset_idcart/val/images
!mkdir /content/yolov9/final_dataset_idcart/val/labels
images_train = glob('/content/yolov9/final_dataset_idcart/images/train/*.jpg')
for img in images_train:
  shutil.move(img,'/content/yolov9/final_dataset_idcart/train/images')
images_test = glob('/content/yolov9/final_dataset_idcart/images/test/*.jpg')
for img in images_test:
  shutil.move(img,'/content/yolov9/final_dataset_idcart/test/images')
images_valid = glob('/content/yolov9/final_dataset_idcart/images/val/*.jpg')
for img in images_valid:
  shutil.move(img,'/content/yolov9/final_dataset_idcart/val/images')
##################################
label_train = glob('/content/yolov9/final_dataset_idcart/labels/train/*.txt')
for label in label_train:
  shutil.move(label,'/content/yolov9/final_dataset_idcart/train/labels')
label_test = glob('/content/yolov9/final_dataset_idcart/labels/test/*.txt')
for label in label_test:
  shutil.move(label,'/content/yolov9/final_dataset_idcart/test/labels')
label_valid = glob('/content/yolov9/final_dataset_idcart/labels/val/*.txt')
for label in label_valid:
  shutil.move(label,'/content/yolov9/final_dataset_idcart/val/labels')
!rm -rf /content/yolov9/final_dataset_idcart/images
!rm -rf /content/yolov9/final_dataset_idcart/labels
```
